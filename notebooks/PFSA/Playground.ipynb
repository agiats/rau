```python
def compare_predictive_information_definitions(pfsa, max_n=10):
    """Compare two different definitions of predictive information (excess entropy).

    1. Direct definition: E = H(S) - h
       Available directly as pfsa.predictive_information

    2. Sum definition: E = sum_n(h_n - h)
       Requires calculating h_n for different context lengths

    Args:
        pfsa: A PFSA instance
        max_n: Maximum context length to consider

    Returns:
        None (prints and plots results)
    """
    import matplotlib.pyplot as plt

    # Check if the PFSA is unifilar (required for predictive information formula)
    is_unifilar = pfsa.is_unifilar()
    print(f"PFSA is unifilar: {is_unifilar}")

    if not is_unifilar:
        print("Warning: PFSA is not unifilar. Predictive information formulas may not be accurate.")

    # Definition 1: E = H(S) - h
    direct_pi = pfsa.predictive_information
    state_entropy = pfsa.state_entropy
    next_symbol_entropy = pfsa.next_symbol_entropy

    print(f"\nDefinition 1: E = H(S) - h")
    print(f"State entropy H(S): {state_entropy:.6f}")
    print(f"Asymptotic entropy rate h: {next_symbol_entropy:.6f}")
    print(f"Predictive information (direct): {direct_pi:.6f}")

    # Definition 2: E = sum_n(h_n - h)
    print(f"\nDefinition 2: E = sum_n(h_n - h)")

    # Calculate h_1 (unigram entropy)
    h_1 = pfsa.unigram_entropy()

    # Calculate h_n for n = 2 to max_n
    h_n_values = [h_1]
    for n in range(2, max_n + 1):
        h_n = pfsa.local_entropy(n)
        h_n_values.append(h_n)

    # Print h_n values and differences from h
    print(f"Asymptotic entropy rate h: {next_symbol_entropy:.6f}")
    sum_diff = 0
    differences = []

    for n, h_n in enumerate(h_n_values, 1):
        diff = h_n - next_symbol_entropy
        sum_diff += diff
        differences.append(diff)
        print(f"  h_{n}: {h_n:.6f} (difference from h: {diff:.6f})")

    print(f"\nPredictive information (sum): {sum_diff:.6f}")
    print(f"Difference between methods: {abs(direct_pi - sum_diff):.6f}")

    # Plot the results
    plt.figure(figsize=(12, 8))

    # Plot 1: h_n values
    plt.subplot(2, 1, 1)
    plt.plot(range(1, len(h_n_values) + 1), h_n_values, 'o-', label='h_n')
    plt.axhline(y=next_symbol_entropy, color='r', linestyle='--', label='h (asymptotic)')
    plt.xlabel('Context Length n')
    plt.ylabel('Conditional Entropy')
    plt.title('Convergence of Conditional Entropy Rates')
    plt.legend()
    plt.grid(True)

    # Plot 2: Cumulative sum of differences
    plt.subplot(2, 1, 2)
    cumsum = [sum(differences[:i+1]) for i in range(len(differences))]
    plt.bar(range(1, len(differences) + 1), differences, alpha=0.5, label='h_n - h')
    plt.plot(range(1, len(cumsum) + 1), cumsum, 'r-o', label='E = sum(h_n - h)')
    plt.axhline(y=direct_pi, color='g', linestyle='--', label='E = H(S) - h')
    plt.xlabel('Context Length n')
    plt.ylabel('Contribution to Predictive Information')
    plt.title('Components of Predictive Information')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    return {
        'direct_pi': direct_pi,
        'sum_pi': sum_diff,
        'h_n_values': h_n_values,
        'differences': differences
    }


results = compare_predictive_information_definitions(A, max_n=6)
```

```python
# pfsa_info.py
"""
Compute all information–theoretic quantities needed for XXX
given a PFSA instance (see PFSA class in the prompt).
"""

from __future__ import annotations

import numpy as np
from numpy.typing import NDArray
from typing import Dict

try:
    import scipy.linalg as la
except ImportError:
    la = None  # fallback will avoid logm if SciPy is missing


def harmonic_fundamental_matrix(
    T: NDArray,
    *,
    tol: float = 1e-12,
    max_terms: int = 2_000,
) -> NDArray:
    """
    Return H = sum_{n>=0} T^n / (n+1).
    Uses logm when possible, otherwise a power-series fallback.
    """
    I = np.eye(T.shape[0])
    # fast path ────────────────────────────────────────────────────────────
    if la is not None:
        try:
            H = -np.linalg.inv(T) @ la.logm(I - T)
            # logm can introduce tiny imaginary parts → drop them
            return H.real
        except (np.linalg.LinAlgError, la.LinAlgError):
            # singular or ill-conditioned T → fall through to series
            pass

    # slow but always safe fallback ───────────────────────────────────────
    H = np.zeros_like(T, dtype=float)
    power = I.copy()
    for n in range(max_terms):
        if n:                      # skip multiplication for n=0
            power = power @ T
        term = power / (n + 1)
        H += term
        if np.max(np.abs(term)) < tol:
            break
    else:
        raise RuntimeError("power-series for H did not converge")
    return H


def build_symbol_kernels(pfsa) -> Dict[str, NDArray]:
    """
    Returns:
        denom  : (n_states,)   outbound prob mass per state (excl. EOS)
        P      : (n_states, n_states, n_symbols)  conditional kernels
        A      : (n_states, n_states)  aggregated kernel
        g      : (n_states,)   local symbol entropy g_i
    """
    n, m = pfsa.n_states, pfsa.n_symbols
    T_syms = np.stack([pfsa.Ts[y] for y in range(m)], axis=-1)  # (n,n,m)
    denom = T_syms.sum(axis=(1, 2))                             # (n,)

    # Avoid divide-by-zero by broadcasting a mask
    safe_denom = denom.copy()
    safe_denom[safe_denom == 0.0] = 1.0
    P = T_syms / safe_denom[:, None, None]                      # (n,n,m)
    P[denom == 0.0] = 0.0

    A = P.sum(axis=-1)                                          # (n,n)

    # local symbol entropy g_i = −Σ_{sym,j} P_ij^sym log P_ij^sym
    with np.errstate(divide="ignore"):
        g = -(P * np.log(P.clip(min=1e-300))).sum(axis=(1, 2))
    g[denom == 0.0] = 0.0                                       # absorbing rows

    return dict(denom=denom, P=P, A=A, g=g)


def pfsa_information(pfsa) -> Dict[str, float | NDArray]:
    """
    Compute XXX and supporting quantities for a *probabilistic* PFSA.
    Raises AssertionError if PFSA is not normalised.
    """
    assert pfsa.is_probabilistic, "PFSA transition rows do not sum to 1"

    λ, ρ, T = pfsa.λ, pfsa.ρ, pfsa.T
    H = harmonic_fundamental_matrix(T)                          # (n,n)
    Z = float(λ @ H @ ρ)                                        # scalar

    # πᵀ = λᵀ H / Z
    π = (λ @ H) / Z                                             # (n,)

    # build P, A, g
    kernels = build_symbol_kernels(pfsa)
    A, g = kernels["A"], kernels["g"]

    # solve (I − A) h = g
    I = np.eye(pfsa.n_states)
    h = np.linalg.solve(I - A, g)                               # (n,)

    # unconditional suffix entropy H(C)
    r = λ @ H                                                   # (n,)
    with np.errstate(divide="ignore"):
        HC = -(1 / Z) * np.dot(r * np.log(r.clip(min=1e-300)), ρ)

    # conditional suffix entropy H(C | Y) = πᵀ h
    H_cond = float(π @ h)

    XXX = HC - H_cond

    # some extras for sanity checks
    mean_length = (λ @ H @ np.ones(pfsa.n_states)) - 1
    next_symbol_entropy = pfsa.next_symbol_entropy

    return dict(
        H=H,
        Z=Z,
        pi=π,
        A=A,
        g=g,
        h=h,
        HC=HC,
        HC_cond=H_cond,
        XXX=XXX,
        mean_length=mean_length,
        next_symbol_entropy=next_symbol_entropy,
    )
```
