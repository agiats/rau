```python
# 必要なインポート
import os
from pathlib import Path
```

```python
input_file = (
    "../results/lstm_results/variations_wc_min1_max20_10K/0002002/seed0/test.txt"
)
```

```python
# ファイルの読み込みと処理
with open(input_file, "r") as file:
    all_lines = file.readlines()

all_lines
```

```python
all_lines = [l for l in all_lines if len(l.split("|")) > 3]
all_lines
```

```python
word_score_lines = [
    l.split("|")[3][1:] for l in all_lines if l.split("|")[3].split(" ")[1].isnumeric()
]
word_score_lines
```

```python
word_score_lines.sort(key=lambda a: int(a.split(" ")[0]))
word_score_lines
```

```python
# テキストの処理
full_text = ""
for line in word_score_lines:
    full_text += " ".join(line.split(" ")[1:]).strip("\n") + " "
full_text
```

```python
# 文章の分割と処理
sentences = []
start = 0
words = full_text.split("\t")
for i in range(len(words)):
    if words[i].split(" ")[0] == "</s>":
        sentences.append(words[start : i + 1])
        start = i + 1
```

```python
sentences[0]
```

```python
def total_sentence_score(words):
    score = 0
    for w in words:
        print(w)
        score += float(w.split(" ")[1][1:-1])
    return score
```

```python
# スコアの計算
sentence_scores = []
for s in sentences:
    sentence_scores.append(total_sentence_score(s))
```
